# Multinomial Logit — Final (Aligned with OLS)
This notebook aligns covariates with the OLS spec, fits MNLogit, reports odds ratios,
derives **High Growth vs Stressed** if needed, and exports ORs + marginal effects to Excel.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from IPython.display import display
```

```python
# === LOAD DATA ===
FILE = "2000 Analysis Python CHANGED.xlsx"
df = pd.read_excel(FILE)
print("Loaded:", df.shape)
```

```python
# === TARGET & CORE CONTINUOUS ===
target_col = "SGrowth_2"
continuous = ['Starting_INT','Starting_BS_Strength','Starting_Size_ln','Starting_Profit']
assert all(c in df.columns for c in continuous+[target_col]), "Missing required columns." 
```

```python
# === DUMMIES ALIGNED WITH OLS ===
# 1) Foreign dummy from P/F (reference = domestic 'P')
df['Foreign'] = (df['P/F'].astype(str).str.upper().str.strip() == 'F').astype(int)

# 2) HGX Increased Export Intensity (reference = 'Decreased Export Intensity')
df['HGX_Increased Export Intensity'] = (df['HGX'].astype(str).str.strip() == 'Increased Export Intensity').astype(int)

# 3) Sector dummies (reference = 'budownictwo')
sector_ref = 'budownictwo'
sector_d = pd.get_dummies(df['Sector'], prefix='Sector', drop_first=False)
ref_col = f"Sector_{sector_ref}"
if ref_col in sector_d.columns:
    sector_d = sector_d.drop(columns=[ref_col])  # drop reference
```

```python
# === DESIGN MATRIX ===
X = pd.concat([
    df[continuous],
    df[['Foreign','HGX_Increased Export Intensity']],
    sector_d
], axis=1)

# Interaction as in OLS
X['Starting_INT_Mult_Starting_Size_ln'] = X['Starting_INT'] * X['Starting_Size_ln']

# Align rows with target
cat_map = {'0.Stressed':0,'1.Declining':1,'2.Stable':2,'3.High Growth':3}
y = df[target_col].map(cat_map)
mask = (~y.isna()) & (~X.isna().any(axis=1))
X = X.loc[mask].copy()
y = y.loc[mask].astype(int)
print("Model N:", X.shape[0], "Features:", X.shape[1])
```

```python
# === FIT MULTINOMIAL LOGIT ===
Xc = sm.add_constant(X, has_constant='add')
mnl = sm.MNLogit(y, Xc).fit()
display(mnl.summary())
```

```python
# === DETECT LAYOUT & BASE, THEN BUILD OR TABLES ===
params = mnl.params
pvals  = mnl.pvalues
bse    = mnl.bse

# If params is DataFrame with columns as categories (common): columns are non-base cats vs base
print("Params shape:", params.shape)
print("Params columns (non-base categories):", list(params.columns))

or_tables = {}
if isinstance(params, pd.DataFrame):
    # Each column k is 'k vs base'
    for k in params.columns.tolist():
        beta  = params[k]
        se    = bse[k]
        lower = beta - 1.96*se
        upper = beta + 1.96*se
        out = pd.DataFrame({
            'Odds Ratio': np.exp(beta),
            'CI Lower':   np.exp(lower),
            'CI Upper':   np.exp(upper),
            'p-value':    pvals[k]
        })
        or_tables[k] = out
else:
    # Series with MultiIndex on rows: (cat, predictor)
    cats = params.index.get_level_values(0).unique().tolist()
    for k in cats:
        beta  = params.loc[k]
        se    = bse.loc[k]
        lower = beta - 1.96*se
        upper = beta + 1.96*se
        out = pd.DataFrame({
            'Odds Ratio': np.exp(beta),
            'CI Lower':   np.exp(lower),
            'CI Upper':   np.exp(upper),
            'p-value':    pvals.loc[k]
        })
        or_tables[k] = out

# Display quick peeks
for k in or_tables:
    print(f"\n### Category {k} vs (base)")
    display(or_tables[k].sort_values('p-value').head(20))
```

```python
# === DERIVE High Growth (3) vs Stressed (0) WHEN BASE IS 3 ===
# If columns are [0,1,2], it's '0/1/2 vs base=3'; derive 3 vs 0 by negating column 0.
derived_hg_vs_0 = None
if isinstance(params, pd.DataFrame) and (0 in params.columns.tolist()) and (3 not in params.columns.tolist()):
    beta = -params[0]          # negate Stressed vs HG
    se   =  bse[0]             # same SE for difference-of-zero-mean contrast
    lower = beta - 1.96*se
    upper = beta + 1.96*se
    derived_hg_vs_0 = pd.DataFrame({
        'Odds Ratio': np.exp(beta),
        'CI Lower':   np.exp(lower),
        'CI Upper':   np.exp(upper),
        'p-value (from Stressed vs HG)': pvals[0]
    })
    print("\n### Derived: 3.High Growth vs 0.Stressed")
    display(derived_hg_vs_0.sort_values('p-value (from Stressed vs HG)').head(20))
```

```python
# === MARGINAL EFFECTS (Average dydx) ===
mfx = mnl.get_margeff(at='overall', method='dydx').summary_frame()
display(mfx.head(40))
```

```python
# === EXPORT TO EXCEL ===
with pd.ExcelWriter("MNLogit_Final_Aligned.xlsx") as w:
    for k in or_tables:
        sheet = f"OR_cat{k}_vs_base"
        or_tables[k].to_excel(w, sheet_name=sheet)
    if derived_hg_vs_0 is not None:
        derived_hg_vs_0.to_excel(w, sheet_name="3_vs_0_Derived")
    mfx.to_excel(w, sheet_name="MarginalEffects_overall")
print("✅ Exported to MNLogit_Final_Aligned.xlsx")
```

Notes:
- Covariates aligned with OLS: sectors (ref: *budownictwo*), Foreign, HGX Increased, INT×Size.
- If base is 3 (High Growth), you will see columns [0,1,2]. The notebook derives 3 vs 0 as well.
- Use the Excel export for clean tables in slides/reports.
