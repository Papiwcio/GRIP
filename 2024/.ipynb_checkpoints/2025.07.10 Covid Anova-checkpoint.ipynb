# DATA PREPARATION

```python
# INITIATION: PYTHON AND DISPLAY

# Import libraries
import pandas as pd #
import numpy as np
import ipywidgets as widgets
import matplotlib.pyplot as plt #charts
import seaborn as sns #charts
from packages.pivot_utils import create_pivot_ui #This is my own customised Pivot Table Function

# For better visibility
from IPython.display import display, HTML

# Apply a no-wrap style to prevent text wrapping
display(HTML("""
<style>
.dataframe td { white-space: nowrap; }
</style>
"""))

# Set the option to display all columns
pd.set_option('display.max_columns', None)

# Set the option to display all rows
pd.set_option('display.max_rows', None)

# Set the float format to use a space as a thousands separator and a comma as a decimal separator
pd.set_option('display.float_format', lambda x: f'{x:,.2f}'.replace(',', ' ').replace('.', ','))
```

```python
# FILE LOAD

# Read the Excel file into a DataFrame
df = pd.read_excel('2000 Analysis Python CHANGED.xlsx', decimal=',')
display (df.head())
```

# DESCRIPTIVE ANALYSIS

```python
# STEP 1: Filter for 2023 only (already done earlier, included here for completeness)
df_2023 = df[df['Year'] == 2023].copy()

# STEP 2: Select and clean required columns (drop missing values)
df_2023_partial = df_2023[['SGrowth_2023_2019', 'RSGrowth_2023_2019', 'HGX', 'HGX%', 'P/F', 'Sales']].dropna().copy()

# STEP 3: Create strategic group labels (Ownership + Export trend)
df_2023_partial['Group'] = df_2023_partial['P/F'] + " - " + df_2023_partial['HGX'].str.replace("Export Intensity", "Export")

# STEP 4: Compute unweighted descriptive statistics (firm-level view)
summary = df_2023_partial.groupby('Group').agg(
    Number_of_Companies=('SGrowth_2023_2019', 'count'),
    Average_Sales_Growth=('SGrowth_2023_2019', 'mean'),
    Average_Real_Sales_Growth=('RSGrowth_2023_2019', 'mean'),
    Average_HGX_Percent=('HGX%', 'mean')
).round(2)

# STEP 5: Format group names for clarity
summary.index = summary.index.str.replace('Export Intensity', 'Export', regex=False)

# STEP 6: Sort groups logically for presentation
group_order = [
    'P - Increased Export', 'P - Decreased Export',
    'F - Increased Export', 'F - Decreased Export'
]
summary = summary.loc[group_order]

# STEP 7: Rename columns for final clean presentation
summary.columns = ['Number of Companies','Average Sales Growth %','Average Real Sales Growth','Average HGX %']

# DISPLAY or EXPORT
display(summary)
summary.to_excel('firm_growth_summary.xlsx')
```

```python
from scipy.stats import f_oneway

# Make sure Group and SGrowth_2023_2019 exist and are clean
grouped_data = df_2023_partial.groupby('Group')['SGrowth_2023_2019'].apply(list)

# Run one-way ANOVA Analysis of Variance
anova_result = f_oneway(*grouped_data)

# Print the result
print("Analysis of Variance (ANOVA) result:")
print(f"F-statistic = {anova_result.statistic:.3f}")
print(f"p-value     = {anova_result.pvalue:.4f}")
```

```python
# Create winsorised version of SGrowth_2023_2019 (clip 1st–99th percentile)
lower = df_2023_partial['SGrowth_2023_2019'].quantile(0.01)
upper = df_2023_partial['SGrowth_2023_2019'].quantile(0.99)

df_2023_partial['SGrowth_2023_2019_W'] = df_2023_partial['SGrowth_2023_2019'].clip(lower, upper)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define group order
group_order = [
    'P - Increased Export', 'P - Decreased Export',
    'F - Increased Export', 'F - Decreased Export'
]

# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(
    data=df_2023_partial,
    x='Group',
    y='SGrowth_2023_2019_W',
    order=group_order
)

plt.title("Boxplot: Sales Growth (2019–2023) by Strategy Group")
plt.ylabel("Sales Growth % (Winsorised)")
plt.xlabel("")
plt.xticks(rotation=15)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig("Results/boxplot_growth.png")
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(
    data=df_2023_partial,
    x='Group',
    y='SGrowth_2023_2019_W',
    order=group_order,
    inner='quartile'
)

plt.title("Violin Plot: Sales Growth Distribution by Strategy Group")
plt.ylabel("Sales Growth % (Winsorised)")
plt.xlabel("")
plt.xticks(rotation=15)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig("Results/violin_growth.png")
plt.show()
```

```python
import pandas as pd
import xlsxwriter

# File paths
excel_path = "Results/growth_analysis_with_charts.xlsx"
boxplot_path = "Results/boxplot_growth.png"
violinplot_path = "Results/violin_growth.png"

# Create ANOVA DataFrame (if needed)
anova_table = pd.DataFrame({
    'F-statistic': [anova_result.statistic],
    'p-value': [anova_result.pvalue]
}).round(4)

# Write to Excel using xlsxwriter
with pd.ExcelWriter(excel_path, engine='xlsxwriter') as writer:
    # Write tables
    summary.to_excel(writer, sheet_name='Descriptive Summary')
    anova_table.to_excel(writer, sheet_name='ANOVA Result', index=False)

    # Access the workbook and worksheet
    workbook  = writer.book
    worksheet = workbook.add_worksheet('Charts')
    writer.sheets['Charts'] = worksheet

    # Insert charts
    worksheet.insert_image('B2', boxplot_path)
    worksheet.insert_image('B25', violinplot_path)

print(f"✅ Excel file with embedded charts saved: {excel_path}")
```
