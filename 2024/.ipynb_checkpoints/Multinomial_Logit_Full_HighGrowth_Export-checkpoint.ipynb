# Multinomial Logistic Regression – High Growth vs Stressed
This notebook fits a multinomial logistic regression on SGrowth_2 and robustly extracts
High Growth (vs Stressed) odds ratios for display and Excel export.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from IPython.display import display
```

```python
# === LOAD FINAL DATA ===
file_path = "2000 Analysis Python CHANGED.xlsx"
df = pd.read_excel(file_path)
print(f"Loaded shape: {df.shape}")
```

```python
# === FILTER YEAR ===
df_2023 = df[df['Year'] == 2023].copy()
print(f"Rows in 2023: {df_2023.shape[0]}")
```

```python
# === DEFINE TARGET AND PREDICTORS ===
predictors = ['Starting_INT', 'Starting_BS_Strength', 'Starting_Size_ln', 'Starting_Profit', 'HGX%']
required_cols = ['SGrowth_2'] + predictors
missing = [c for c in required_cols if c not in df_2023.columns]
if missing:
    raise KeyError(f"Missing required columns in data: {missing}")
df_model = df_2023[required_cols].dropna().copy()
print(f"Modelling rows after dropna: {df_model.shape[0]}")
```

```python
# === MAP CATEGORIES TO NUMERIC ===
category_map = {
    '0.Stressed': 0,
    '1.Declining': 1,
    '2.Stable': 2,
    '3.High Growth': 3
}
df_model['SGrowth_2_num'] = df_model['SGrowth_2'].map(category_map)
if df_model['SGrowth_2_num'].isna().any():
    unmapped = df_model.loc[df_model['SGrowth_2_num'].isna(), 'SGrowth_2'].unique().tolist()
    raise ValueError(f"Unmapped SGrowth_2 labels found: {unmapped}")
```

```python
# === FIT MULTINOMIAL LOGISTIC REGRESSION ===
X = sm.add_constant(df_model[predictors])
y = df_model['SGrowth_2_num']
model = sm.MNLogit(y, X)
result = model.fit()
```

```python
# === SUMMARY ===
display(result.summary())
```

```python
# === DISPLAY AND EXPORT HIGH GROWTH ODDS RATIOS (robust) ===
params = result.params
conf = result.conf_int()

# Inspect available categories
cat_keys = params.index.get_level_values(0).unique().tolist()
print("Detected outcome categories (vs. base):", cat_keys)

# Pick the 'High Growth' key robustly
high_growth_key = None
# Prefer numeric 3 (consistent with mapping) if present
if any((isinstance(k, (int, np.integer)) and k == 3) for k in cat_keys):
    high_growth_key = 3
# Otherwise, look for a string containing 'High Growth'
elif any(('High Growth' in str(k)) for k in cat_keys):
    high_growth_key = next(k for k in cat_keys if 'High Growth' in str(k))
# Finally, try literal string '3'
elif any((str(k) == '3') for k in cat_keys):
    high_growth_key = next(k for k in cat_keys if str(k) == '3')

if high_growth_key is None:
    raise KeyError(
        "Could not identify the 'High Growth' category. "
        f"Available categories are: {cat_keys}. "
        "If your label differs, adjust the detection logic."
    )

# Build odds-ratio table safely
conf_hi = conf.loc[high_growth_key]
odds_high_growth = pd.DataFrame({
    'Odds Ratio': np.exp(params.loc[high_growth_key]),
    'CI Lower':   np.exp(conf_hi.iloc[:, 0]),
    'CI Upper':   np.exp(conf_hi.iloc[:, 1]),
})

# Display in notebook
display(odds_high_growth)

# Export to Excel
odds_high_growth.to_excel("High_Growth_vs_Stressed_Odds.xlsx")
print("✅ Results exported to High_Growth_vs_Stressed_Odds.xlsx")
```

End of notebook.
