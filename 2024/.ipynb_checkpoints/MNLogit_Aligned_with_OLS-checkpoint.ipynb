# Multinomial Logit aligned with OLS spec
This notebook rebuilds the OLS feature set (sector dummies, Foreign, HGX change, INT×Size),
fits a multinomial logit with base = 0.Stressed, then exports odds ratios and marginal effects.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from IPython.display import display
```

```python
# === LOAD DATA ===
FILE = "2000 Analysis Python CHANGED.xlsx"
df = pd.read_excel(FILE)
print(df.shape)
```

```python
# === TARGET AND CORE CONTINUOUS ===
target_col = "SGrowth_2"
continuous = ['Starting_INT','Starting_BS_Strength','Starting_Size_ln','Starting_Profit']  # same cores as OLS
assert all(c in df.columns for c in continuous+[target_col])
```

```python
# === REBUILD OLS DUMMIES ===
# 1) Foreign dummy from P/F (reference P)
df['Foreign'] = (df['P/F'].astype(str).str.upper().str.strip() == 'F').astype(int)

# 2) HGX increased dummy (reference 'Decreased Export Intensity')
df['HGX_Increased Export Intensity'] = (df['HGX'].astype(str).str.strip() == 'Increased Export Intensity').astype(int)

# 3) Sector dummies with reference 'budownictwo'
sector_ref = 'budownictwo'
sector_d = pd.get_dummies(df['Sector'], prefix='Sector')
drop_col = f"Sector_{sector_ref}"
if drop_col in sector_d.columns:
    sector_d = sector_d.drop(columns=[drop_col])  # drop reference

# Combine design matrix
X = pd.concat([df[continuous], df[['Foreign','HGX_Increased Export Intensity']], sector_d], axis=1)
```

```python
# === INTERACTION: Starting_INT * Starting_Size_ln (as in OLS) ===
X['Starting_INT_Mult_Starting_Size_ln'] = X['Starting_INT'] * X['Starting_Size_ln']
```

```python
# === ALIGN ROWS: Need rows where y and all X are present ===
cat_map = {'0.Stressed':0,'1.Declining':1,'2.Stable':2,'3.High Growth':3}
y = df[target_col].map(cat_map)
mask = (~y.isna()) & (~X.isna().any(axis=1))
y = y[mask].astype(int)
X = X.loc[mask].copy()
print("Model N:", X.shape[0], "Features:", X.shape[1])
```

```python
# === FIT MULTINOMIAL (base = 0 by construction) ===
X_const = sm.add_constant(X, has_constant='add')
mnl = sm.MNLogit(y, X_const).fit()
display(mnl.summary())
```

```python
# === ODDS RATIOS WITH 95% CI PER CATEGORY (vs base 0.Stressed) ===
params = mnl.params     # rows: category (1,2,3), cols: predictors (const + X)
bse = mnl.bse
or_df = {}
for cat in params.index:  # 1,2,3
    beta = params.loc[cat]
    se   = bse.loc[cat]
    lower = beta - 1.96*se
    upper = beta + 1.96*se
    out = pd.DataFrame({
        'Odds Ratio': np.exp(beta),
        'CI Lower':   np.exp(lower),
        'CI Upper':   np.exp(upper),
        'p-value':    mnl.pvalues.loc[cat]
    })
    or_df[int(cat)] = out

# Display one by one
for k in sorted(or_df):
    print(f"\n### Category {k} vs 0.Stressed")
    display(or_df[k].sort_index())
```

```python
# === MARGINAL EFFECTS (Average) ===
mfx = mnl.get_margeff(at='overall', method='dydx').summary_frame()
display(mfx.head(30))
```

```python
# === EXPORT EXCEL ONE-PAGER ===
with pd.ExcelWriter("MNLogit_Aligned_with_OLS.xlsx") as w:
    for k in sorted(or_df):
        or_df[k].to_excel(w, sheet_name=f"OR_cat{k}_vs_0")
    mfx.to_excel(w, sheet_name="MarginalEffects_overall")
print("✅ Exported to MNLogit_Aligned_with_OLS.xlsx")
```

Notes:
- Base category is **0.Stressed**.
- Odds ratios > 1 mean higher odds of being in category k vs Stressed as the predictor increases.
- Review sector dummies to ensure the reference sector matches OLS (`budownictwo`).
