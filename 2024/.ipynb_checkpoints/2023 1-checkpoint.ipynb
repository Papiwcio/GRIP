# DATA PREPARATION

```python
# INITIATION: PYTHON AND DISPLAY

# Import libraries
import pandas as pd #
import numpy as np

# For better visibility
from IPython.display import display, HTML

# Apply a no-wrap style to prevent text wrapping
display(HTML("""
<style>
.dataframe td { white-space: nowrap; }
</style>
"""))

# Set the option to display all columns
pd.set_option('display.max_columns', None)

# Set the option to display all rows
pd.set_option('display.max_rows', None)

# Set the float format to use a space as a thousands separator and a comma as a decimal separator
pd.set_option('display.float_format', lambda x: f'{x:,.2f}'.replace(',', ' ').replace('.', ','))
```

```python
# FILE LOAD

# Read the Excel file and specify that 'PKD' should be read as a string to preserve leading zeros
df = pd.read_excel('2000 Analysis Python.xlsx', sheet_name='2023', decimal=',', dtype={'PKD': str})

pkd = pd.read_excel('2000 PKD.xlsx')  # Load the pkd data

# Convert 'PKD' to string and extract the first two characters (keeping leading zeros)
df['PKD_2'] = df['PKD'].astype(str).str.zfill(2).str[:2]

# Convert the 'PKD_2' column in pkd to a string to match the type of 'PKD'
pkd['PKD_2'] = pkd['PKD_2'].astype(str).str.zfill(2)

# Perform the merge using 'PKD' and 'Number'
df = pd.merge(
    df,
    pkd,
    left_on='PKD_2',
    right_on='PKD_2',
    how='left'
)

cols = df.columns.tolist()  # Get a list of all column names
cols.insert(25, cols.pop(cols.index('PKD_2'))) # Remove 'PKD_2' and insert it at index 25
df = df[cols]  # Reassign the DataFrame with the new column order

df.columns = df.columns.astype(str)
```

```python
# FILE INSPECTION

# Print the number of rows and columns
print(f"Number of rows and columns: {df.shape}")

# Display the first 5 rows of the DataFrame
display(df.head())

# Print the last 5 rows of the DataFrame
display(df.tail())
```

```python
# DATA CLEAN-UP

# Step 1: Identify duplicates based on 'Name' and 'Year' columns
duplicates = df[df.duplicated(subset=['Name', 'Year'], keep=False)]

# Step 2: Display the duplicates for inspection
print("Duplicate rows based on 'Name' and 'Year':")
print(duplicates)
```

```python
#DATA TYPES ASSIGNMENT

# Step 1: Changing data types for analysis
dtype_changes = {
    
    'PKD': 'category',
    'PKD_2': 'category',
    'PKD_1': 'category',
    'Owner Type': 'category',
    'Year': 'category',
    'Sales': 'float',
    'Gross Profit': 'float',
    'Operating Profit': 'float',
    'Net Profit': 'float',
    'Depreciation': 'float',
    'Export Sales': 'float',
    'Employment': 'float',
    'Staff costs': 'float',
    'BS': 'float',
    'Operating Profit': 'float',
    'Fixed Assets': 'float',
    'Current Assets': 'float',
    'Equity': 'float',
    'Liabilities': 'float',
    'Quartile of digital intensity: 2001-03': 'category',
    'Quartile of digital intensity: 2013-15': 'category'
    
    
    
    }
# Step 2: Check each column individually to identify possible issues

for col, dtype in dtype_changes.items():
    try:
        if dtype in ['float', 'int']:  # Check if the target type is numeric
            df[col] = pd.to_numeric(df[col], errors='coerce')  # Use pd.to_numeric for numeric conversion
        df[col] = df[col].astype(dtype)  # Convert to the target dtype
        print(f"Successfully converted {col} to {dtype}")
    except ValueError as e:
        print(f"Error converting {col} to {dtype}: {e}")
```

```python
display(df.info())
```

```python
# Save integrated file
df.to_excel('2000 Analysis Python INTEGRATED.xlsx', index=False)

print("DataFrame 'df' has been saved to '2000 Analysis Python INTEGRATED.xlsx'.")
```

# ANALYSIS  SET-UP

```python
#SET THE MAIN FILTERS FOR THE ANALYSIS

#filter companies
company_filter = df[df['PKD_2'] == '19'] #filter out fuel companies

# Identify the names with Sales = 0 or Employment = 0 in 2019 or 2022
names_to_remove = df[
    (df['Year'].isin([2019, 2023])) & 
    (
        (df['Sales'] == 0) | pd.isnull(df['Sales']) |
        (df['Employment'] == 0) | pd.isnull(df['Employment'])
    )
]['Name'].unique() 

# Apply the filters (companies on the list in 2019 and missing data from above calculation)
df_filter = df[
    df['2019'].notna() &
    (~df['Name'].isin(names_to_remove))
]

#Filter our selected PKD2
df_filter = df_filter[~df_filter.index.isin(company_filter.index)]

# Show results

print(f"Number of companies removed due to sector considerations is: {company_filter['Name'].nunique()}")
print(f"Number of companies with sales or employment in 2019 or 2023 equal to 0 is: {len(names_to_remove)}")
print(f"Number of companies, which did not appear on the 2000 List in 2023 is: {len(df[df['2023'].isna()])}")

print(f"Number of companies after clean-up: {len(df_filter[df_filter['Year'] == 2022]['Sales'])}")
```

```python
#Lista firm usuniętych z analizy
df_print = df[df['Name'].isin(names_to_remove)]

display(df_print.pivot(index='Name', columns='Year', values='Employment'))
df_print.to_excel('filtered_data.xlsx', index=False)  # Save the filtered DataFrame to an Excel file
```

```python
#CHECK OF FILTERING

# Select the columns you want to include in the groupby and describe operations
selected_columns = ['Sales', 'Employment']

# Group by Year' and apply describe to the selected columns only
df_summary = df.groupby(['Year'], observed=False)[selected_columns].describe().transpose()
df_filter_summary = df_filter.groupby(['Year'], observed=False)[selected_columns].describe().transpose()
display(df_summary, df_filter_summary)

# Show max and min company
selected_columns = ['Name','Sales', 'Employment']
display(df_filter['Name'].nunique())
display(df_filter.loc[df_filter['Sales'].idxmax(),selected_columns])
display(df_filter.loc[df_filter[df_filter['Year'] == 2023]['Sales'].idxmin(),selected_columns])

display(df_filter[selected_columns].head())

display(df_filter[selected_columns].tail())
```

```python
#ADDING ANALYSIS DIMENSIONS

# Create a copy of df_filter to avoid the SettingWithCopyWarning
df_filter = df_filter.copy()

#Export as % of Sales
df_filter['Export as % of Sales']= df_filter['Export Sales'] / df_filter['Sales']

#ROA / ROS / ROE
df_filter['ROS']= df_filter['Net Profit'] / df_filter['Sales']
df_filter['ROA']= df_filter['Net Profit'] / df_filter['BS']  
df_filter['ROE']= df_filter['Net Profit'] / df_filter['Equity']

# Polish / Foreign Companies
df_filter['P/F'] = np.where((df_filter['Owner Type'].astype(str).str[:1] == '5'), 'F', 'P')
df_filter['P/F'] = df_filter['P/F'].astype('category')

# Starting Profit - ROS in 2019
ros_2019 = df_filter[df_filter['Year'] == 2019].set_index('Name')['ROS'] # Step 1: Extract 2019 ROS values
df_filter['Starting_Profit'] = df_filter['Name'].map(ros_2019)# Step 2: Map 2019 ROS values to the 'Starting_Profit' column for all rows

# Starting BalanceSheet Strength - Equity / BS in 2019
df_filter['BS'] = df_filter['BS'].replace(0, np.nan) # Replace 0 in the BS column with NaN to avoid division by zero

df_filter['BS_Strength'] = df_filter['Equity'] / df_filter['BS']
bs_2019 = df_filter[df_filter['Year'] == 2019].set_index('Name')['BS_Strength'] # Step 1: Extract 2019 BS_Strength values
df_filter['Starting_BS_Strength'] = df_filter['Name'].map(bs_2019)# Step 2: Map 2019 ROS values to the 'Starting_Profit' column for all rows


# Starting Size - ln sales in 2019
ln_sales_2019 = df_filter[df_filter['Year'] == 2019].set_index('Name')['Sales'].apply(np.log)  # Step 1: Calculate log(Sales) for 2019
df_filter['Starting_Size'] = df_filter['Name'].map(ln_sales_2019)  # Step 2: Map these values to the 'Starting_Profit' column

# Human Capital Proxy - Staff costs / employment
df_filter['Human_Capital'] = df_filter['Staff costs'] / df_filter['Employment']

#cols = df_filter.columns.tolist()  # Get a list of all column names
#cols.insert(9, cols.pop(cols.index('P/F'))) # Remove 'P/F' and insert it at index 9
#df_filter = df_filter[cols]  # Reassign the DataFrame with the new column order
```

```python
#SALES
# Isolate the annual growth rate threshold
inflation = 8.5 # (38,5% inflation between 2019 and 2023)
stability_threshold = 10 #(97,2% of compound growth between 2019 and 2023)
growth_threshold = 20 #(172,8% of compound growth between 2019 and 2023)

# Pivot the DataFrame to make year-based comparison easier for both Sales and Export
#(use only with 2022 set as a filter)
pivot_sales = df_filter.pivot(index='Name', columns='Year', values='Sales')
pivot_export_sales = df_filter.pivot(index='Name', columns='Year', values='Export Sales')



# High Internationalisation Export Sales > 25% of Sales
pivot_sales['Hint'] = (pivot_export_sales[2023] / pivot_sales[2023]) > 0.25
pivot_sales['Hint'] = pivot_sales['Hint'].apply(lambda x: 'Yes' if x else 'No')

# Sales Growth_2023_2019 (Total Growth)
pivot_sales['SGrowth_2023_2019'] = (pivot_sales[2023] - pivot_sales[2019]) / pivot_sales[2019] * 100
pivot_sales['Annualised_Growth'] = ((pivot_sales[2023] / pivot_sales[2019]) ** (1/4) - 1) * 100
pivot_sales['LSGrowth_2023_2019'] = np.log(pivot_sales[2023] / pivot_sales[2019])


# Stable growth Calculation
pivot_sales['G1'] = ((pivot_sales[2023] / pivot_sales[2019]) ** (1/4) - 1) * 100
pivot_sales['G1'] = pivot_sales['G1'].apply(lambda x: 'Yes' if inflation < x < (stability_threshold + inflation) else 'No')

# Medium Growth Calculation
pivot_sales['G2'] = ((pivot_sales[2023] / pivot_sales[2019]) ** (1/4) - 1) * 100
pivot_sales['G2'] = pivot_sales['G2'].apply(lambda x: 'Yes' if  (stability_threshold + inflation) < x < (growth_threshold+inflation)else 'No')

# High Growth Calculation
pivot_sales['G3'] = ((pivot_sales[2023] / pivot_sales[2019]) ** (1/4) - 1) * 100 > (growth_threshold+inflation)
pivot_sales['G3'] = pivot_sales['G3'].apply(lambda x: 'Yes' if x else 'No')

# Sales Drop Conditions
pivot_sales['Drop_Sales'] = (pivot_sales[2020] < pivot_sales[2019]) | (pivot_sales[2021] < pivot_sales[2020]) | (pivot_sales[2022] < pivot_sales[2021]) | (pivot_sales[2023] < pivot_sales[2022])
pivot_sales['Drop_Sales'] = pivot_sales['Drop_Sales'].apply(lambda x: 'Yes' if x else 'No')

# Add Summary Column
def categorize_growth(row):
    if row['Annualised_Growth'] > (growth_threshold + inflation):
        return 'G3'
    elif (stability_threshold + inflation) < row['Annualised_Growth'] <= (growth_threshold + inflation):
        return 'G2'
    elif inflation < row['Annualised_Growth'] <= (stability_threshold + inflation):
        return 'G1'
    else:
        return 'G0'

pivot_sales['4_Growth_Categories'] = pivot_sales.apply(categorize_growth, axis=1)

# Add 2 Summary Column
def categorize_growth(row):
    if row['Annualised_Growth'] > (5% + inflation):
        return 'G2'
    elif inflation < row['Annualised_Growth'] <= (5% + inflation):
        return 'G1'
    else:
        return 'G0'

pivot_sales['3_Growth_Categories'] = pivot_sales.apply(categorize_growth, axis=1)

# Merge the calculated columns back to the original df_filter DataFrame
df_filter = df_filter.merge(pivot_sales[['SGrowth_2023_2019', 'Annualised_Growth', 'LSGrowth_2023_2019', 'Drop_Sales', 'G1', 'G2', 'G3','4_Growth_Categories','3_Growth_Categories','Hint']], on='Name')
```

```python
#EMPLOYMENT
# Pivot the DataFrame to make year-based comparison easier for Employment
#(use only with 2022 set as a filter)
pivot_employment = df_filter.pivot(index='Name', columns='Year', values='Employment')

# Calculate Employment Growth_2023_2019 (Total Growth)
pivot_employment['EGrowth_2023_2019'] = (pivot_employment[2023] - pivot_employment[2019]) / pivot_employment[2019] * 100
pivot_employment['Annualised_EGrowth'] = ((pivot_employment[2023] / pivot_employment[2019]) ** (1/4) - 1) * 100
pivot_employment['LEGrowth_2023_2019'] = np.log(pivot_employment[2023] / pivot_employment[2019])


# Stable growth calculation
pivot_employment['SE'] = pivot_employment['EGrowth_2023_2019'].apply(lambda x: 'Yes' if x >0 else 'No')

# High Growth Calculation
pivot_employment['HEG'] = pivot_employment['EGrowth_2023_2019'].apply(lambda x: 'Yes' if x >(1.1**4-1) else 'No')

# Add Summary Column
def categorize_egrowth(row):
    if row['EGrowth_2023_2019'] < 0:
        return 'EG0'
    elif (0) < row['EGrowth_2023_2019'] <= (1.1**4-1):
        return 'EG1'
    else:
        return 'EG2'

pivot_employment['EGrowth_Category'] = pivot_employment.apply(categorize_egrowth, axis=1)
    
# Calculate Employment Drop conditions
pivot_employment['Drop_Employment'] = (pivot_employment[2020] < pivot_employment[2019]) | (pivot_employment[2021] < pivot_employment[2020]) | (pivot_employment[2022] < pivot_employment[2021]) | (pivot_employment[2023] < pivot_employment[2022])
pivot_employment['Drop_Employment'] = pivot_employment['Drop_Employment'].apply(lambda x: 'Yes' if x else 'No')


# Step 5: Merge the calculated columns back to the original df_filter DataFrame
df_filter = df_filter.merge(pivot_employment[['EGrowth_2023_2019', 'Drop_Employment','SE','HEG','EGrowth_Category']], on='Name')
```
