```python
# Plotting the original compound growth charts with inflation included as two independent charts
import matplotlib.pyplot as plt
import pandas as pd

# Actual Data provided
data_revenue = {
    "Category": ["Przychody przedsiębiorstw niefinansowych", "Przychody 1500 firm", "Inflacja"],
    "2019": [100, 100, 100],
    "2020": [98.64, 100.89, 103.40],
    "2021": [119.56, 121.25, 108.67],
    "2022": [153.65, 153.24, 124.32],    
    "2023": [160.86, 159.25, 138.50]
}
data_financial_result = {
    "Category": ["Wynik finansowy netto przedsiębiorstw niefinansowych", "Wynik finansowy netto 1500 firm", "Inflacja"],
    "2019": [100, 100, 100],
    "2020": [97.93, 77.54, 103.40],
    "2021": [184.92, 177.38, 108.67],
    "2022": [214.49, 200.77, 124.32],
    "2023": [181.01, 150.77, 138.50]
}

# Creating DataFrames from the data
df_revenue_updated_compound = pd.DataFrame(data_revenue)
df_financial_result_updated_compound = pd.DataFrame(data_financial_result)

# Defining consistent colors for the plots
colors_consistent = ['blue', 'green', 'red']

# Plot 1: Revenue Compound Growth with 2019 as base year
plt.figure(figsize=(10, 6))
for i, row in df_revenue_updated_compound.iterrows():
    plt.plot(
        ["2019", "2020", "2021", "2022","2023"], 
        row[1:], 
        marker='o', 
        color=colors_consistent[i], 
        label=row["Category"],
        linewidth=4  # double thickness
    )
    plt.text(3.05, row["2023"], f'{row["2023"]:.2f}', fontsize=10, color=colors_consistent[i], ha='left', va='bottom')
plt.title("Skumulowany wzrost przychodów przedsiębiorstw (2019-2023, Rok bazowy 2019 = 100)")
plt.xlabel("Rok")
plt.ylabel("Indeks (Skumulowany wzrost, 2019 = 100)")
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.savefig('chart1.png')
plt.show()

# Plot 2: Net Financial Result Compound Growth with 2019 as base year
plt.figure(figsize=(10, 6))
for i, row in df_financial_result_updated_compound.iterrows():
    plt.plot(
        ["2019", "2020", "2021", "2022","2023"], 
        row[1:], 
        marker='o', 
        color=colors_consistent[i], 
        label=row["Category"],
        linewidth=4  # double thickness
    )
    plt.text(3.05, row["2023"], f'{row["2023"]:.2f}', fontsize=10, color=colors_consistent[i], ha='left', va='bottom')
plt.title("Skumulowany wzrost wyniku finansowego netto przedsiębiorstw (2019-2023, Rok bazowy 2019 = 100)")
plt.xlabel("Rok")
plt.ylabel("Indeks (Skumulowany wzrost, 2019 = 100)")
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.savefig('chart2.png')
plt.show()
```
