# Multinomial Logistic Regression – High Growth vs Stressed

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from IPython.display import display
```

```python
# === LOAD FINAL DATA ===
file_path = "2000 Analysis Python CHANGED.xlsx"
df = pd.read_excel(file_path)
```

```python
# === FILTER YEAR ===
df_2023 = df[df['Year'] == 2023].copy()
```

```python
# === DEFINE TARGET AND PREDICTORS ===
predictors = ['Starting_INT', 'Starting_BS_Strength', 'Starting_Size_ln', 'Starting_Profit', 'HGX%']
df_model = df_2023[['SGrowth_2'] + predictors].dropna().copy()
```

```python
# === MAP CATEGORIES TO NUMERIC ===
category_map = {
    '0.Stressed': 0,
    '1.Declining': 1,
    '2.Stable': 2,
    '3.High Growth': 3
}
df_model['SGrowth_2_num'] = df_model['SGrowth_2'].map(category_map)
```

```python
# === FIT MULTINOMIAL LOGISTIC REGRESSION ===
X = sm.add_constant(df_model[predictors])
y = df_model['SGrowth_2_num']
model = sm.MNLogit(y, X)
result = model.fit()
```

```python
# === DISPLAY FULL SUMMARY ===
display(result.summary())
```

```python
# === DISPLAY AND EXPORT HIGH GROWTH ODDS RATIOS ===
params = result.params
conf = result.conf_int()

# Find the correct key for "High Growth" (robust to string labels)
high_growth_key = [k for k in params.index.get_level_values(0).unique() if "High Growth" in str(k)][0]

# Build odds ratio table for High Growth vs Stressed
odds_high_growth = pd.DataFrame({
    'Odds Ratio': np.exp(params.loc[high_growth_key]),
    'CI Lower': np.exp(conf.loc[high_growth_key][0]),
    'CI Upper': np.exp(conf.loc[high_growth_key][1])
})

# Display in notebook
display(odds_high_growth)

# Export to Excel
odds_high_growth.to_excel("High_Growth_vs_Stressed_Odds.xlsx")

print("✅ Results exported to High_Growth_vs_Stressed_Odds.xlsx")
```

This notebook extracts and exports **High Growth vs Stressed** odds ratios to Excel.
