# Multinomial Logistic Regression – High Growth vs Stressed (Robust CI)
This notebook fits MNLogit and **robustly** extracts High Growth vs Stressed odds ratios,
handling both possible statsmodels layouts (MultiIndex rows or category columns).

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from IPython.display import display
```

```python
# === LOAD FINAL DATA ===
file_path = "2000 Analysis Python CHANGED.xlsx"
df = pd.read_excel(file_path)
print(f"Loaded shape: {df.shape}")
```

```python
# === FILTER YEAR ===
df_2023 = df[df['Year'] == 2023].copy()
print(f"Rows in 2023: {df_2023.shape[0]}")
```

```python
# === DEFINE TARGET AND PREDICTORS ===
predictors = ['Starting_INT', 'Starting_BS_Strength', 'Starting_Size_ln', 'Starting_Profit', 'HGX%']
required_cols = ['SGrowth_2'] + predictors
missing = [c for c in required_cols if c not in df_2023.columns]
if missing:
    raise KeyError(f"Missing required columns in data: {missing}")
df_model = df_2023[required_cols].dropna().copy()
print(f"Modelling rows after dropna: {df_model.shape[0]}")
```

```python
# === MAP CATEGORIES TO NUMERIC ===
category_map = {'0.Stressed': 0, '1.Declining': 1, '2.Stable': 2, '3.High Growth': 3}
df_model['SGrowth_2_num'] = df_model['SGrowth_2'].map(category_map)
if df_model['SGrowth_2_num'].isna().any():
    unmapped = df_model.loc[df_model['SGrowth_2_num'].isna(), 'SGrowth_2'].unique().tolist()
    raise ValueError(f"Unmapped SGrowth_2 labels found: {unmapped}")
```

```python
# === FIT MULTINOMIAL LOGISTIC REGRESSION ===
X = sm.add_constant(df_model[predictors])
y = df_model['SGrowth_2_num']
model = sm.MNLogit(y, X)
result = model.fit()
```

```python
# === SUMMARY ===
display(result.summary())
```

```python
params = result.params   # DataFrame: rows=predictors, columns=non-base categories
bse    = result.bse      # same shape as params

# Sanity print
print("Detected non-base category columns:", list(params.columns))

# We assume your mapping: 0=Stressed, 1=Declining, 2=Stable, 3=High Growth
# If columns are [0,1,2], then missing class 3 is the base => High Growth is base.
all_classes = sorted(df_model['SGrowth_2_num'].unique().tolist())
non_base    = list(params.columns)
base_class  = [c for c in all_classes if c not in non_base]
print("Base category detected:", base_class)

if len(base_class) != 1:
    raise ValueError(f"Could not detect a single base class. all={all_classes}, non_base={non_base}, base={base_class}")

base_class = base_class[0]

# We want High Growth (3) vs Stressed (0)
high_growth = 3
stressed    = 0

if base_class == high_growth and stressed in non_base:
    # High Growth is base: use NEGATIVE of Stressed column
    coef = -params[stressed]
    se   =  bse[stressed]
elif base_class == stressed and high_growth in non_base:
    # Stressed is base: use High Growth column directly
    coef = params[high_growth]
    se   = bse[high_growth]
else:
    # More complex contrast (neither is base) – compute β_high - β_stress with covariances
    # Try to use covariance matrix if available
    cov = result.cov_params()
    # Build odds for each predictor via delta method
    diffs = []
    lowers = []
    uppers = []
    idx = params.index  # predictor names
    for p in idx:
        # Identify keys in covariance for (cat, predictor)
        # This path assumes cov has a MultiIndex ((cat, pred), (cat, pred)).
        # If not, you may need to adapt to your statsmodels version.
        key_h = (high_growth, p)
        key_s = (stressed, p)
        beta_h = params.loc[p, high_growth]
        beta_s = params.loc[p, stressed]
        diff = beta_h - beta_s
        # variance of difference
        var = cov.loc[key_h, key_h] + cov.loc[key_s, key_s] - 2 * cov.loc[key_h, key_s]
        se_diff = np.sqrt(var)
        diffs.append(diff)
        lowers.append(diff - 1.96*se_diff)
        uppers.append(diff + 1.96*se_diff)
    odds_high_growth = pd.DataFrame({
        'Odds Ratio': np.exp(diffs),
        'CI Lower':   np.exp(lowers),
        'CI Upper':   np.exp(uppers),
    }, index=idx)
    display(odds_high_growth)
    odds_high_growth.to_excel("High_Growth_vs_Stressed_Odds.xlsx")
    print("✅ Results exported to High_Growth_vs_Stressed_Odds.xlsx")
    raise SystemExit

# 95% CI around the (possibly negated) coef
lower = coef - 1.96*se
upper = coef + 1.96*se

odds_high_growth = pd.DataFrame({
    'Odds Ratio': np.exp(coef),
    'CI Lower':   np.exp(lower),
    'CI Upper':   np.exp(upper),
})

display(odds_high_growth)
odds_high_growth.to_excel("High_Growth_vs_Stressed_Odds.xlsx")
print("✅ Results exported to High_Growth_vs_Stressed_Odds.xlsx")
```

```python
# === EXPORT TO EXCEL ===
with pd.ExcelWriter("Multinomial_Logit_Output.xlsx") as writer:
    result.params.to_excel(writer, sheet_name="Coefficients")
    odds_high_growth.to_excel(writer, sheet_name="Odds Ratios")
```
