```python
# FILE LOAD

# Read the Excel file and specify that 'PKD' should be read as a string to preserve leading zeros
df = pd.read_excel('2000 Analysis Python.xlsx', decimal=',', dtype={'PKD': str})

companies = pd.read_excel('2000 Analysis Python new columns.xlsx')

# Convert the 'PKD_2' column in pkd to a string to match the type of 'PKD'
pkd['PKD_2'] = pkd['PKD_2'].astype(str).str.zfill(2)

# Perform the merge using 'PKD' and 'Number'
df = pd.merge(
    df,
    pkd,
    left_on='PKD_2',
    right_on='PKD_2',
    how='left'
)

cols = df.columns.tolist()  # Get a list of all column names
cols.insert(30, cols.pop(cols.index('PKD_2'))) # Remove 'PKD_2' and insert it at index 30
df = df[cols]  # Reassign the DataFrame with the new column order

# Select the columns and make a copy to avoid the SettingWithCopyWarning
columns_to_select = [5, 0, 1, 2, 3]

companies_subset = companies.iloc[:, columns_to_select].copy()

# Rename the columns with meaningful names
companies_subset.columns = ['NIP', 'L2022', 'L2021', 'L2020', 'L2019']

# Perform the merge
df = pd.merge(
    df,
    companies_subset,
    on='NIP',
    how='left'
)
```
