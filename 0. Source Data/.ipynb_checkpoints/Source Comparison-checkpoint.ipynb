```python
import pandas as pd

# File path
file_path = "Source Tracker.xlsx"

# Read both sheets
df_2019_2024 = pd.read_excel(file_path, sheet_name="2019-2024")
df_2019_2023 = pd.read_excel(file_path, sheet_name="2019-2023")

# Ensure NIP is treated as string (important for identifiers)
df_2019_2024["NIP"] = df_2019_2024["NIP"].astype(str)
df_2019_2023["NIP"] = df_2019_2023["NIP"].astype(str)

# Merge datasets using NIP
comparison = df_2019_2024.merge(
    df_2019_2023,
    on="NIP",
    how="outer",
    indicator=True,
    suffixes=("_2024sheet", "_2023sheet")
)

# Companies present in both sheets
both = comparison[comparison["_merge"] == "both"]

# Only in first sheet
only_2024 = comparison[comparison["_merge"] == "left_only"]

# Only in second sheet
only_2023 = comparison[comparison["_merge"] == "right_only"]

print("In both:", len(both))
print("Only in 2019-2024:", len(only_2024))
print("Only in 2019-2023:", len(only_2023))

# Optional: export results
with pd.ExcelWriter("NIP_comparison.xlsx") as writer:
    both.to_excel(writer, sheet_name="both", index=False)
    only_2024.to_excel(writer, sheet_name="only_2019_2024", index=False)
    only_2023.to_excel(writer, sheet_name="only_2019_2023", index=False)
```
