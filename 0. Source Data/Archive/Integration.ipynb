```python
import pandas as pd #

# FILE LOAD

# Read the Excel
df = pd.read_excel('2000 Analysis Python Old.xlsx', decimal=',',dtype={'NIP': str})

df1 = pd.read_excel('2000 Analysis Python new columns.xlsx', decimal=',',dtype={'NIP': str})

# Extract only the NIP and new columns
df1_subset = df1[['NIP', 'PKD Opis', 'Sector']].copy()

# Merge with left join
df_merged = pd.merge(df, df1_subset, on='NIP', how='left')

# Reorder columns: insert 'Opis' and 'Sector' after 'PKD'
cols = df_merged.columns.tolist()

# Find position of 'PKD'
pkd_index = cols.index('PKD')

# Build new column order
new_order = cols[:pkd_index + 1] + ['PKD Opis', 'Sector'] + [col for col in cols if col not in ('PKD Opis', 'Sector') and col not in cols[:pkd_index + 1]]

# Apply new column order
df = df_merged[new_order]

# Save integrated file
df.to_excel('2000 Analysis Python.xlsx', index=False)
```
