```python
import pandas as pd

df = pd.read_excel("Data_period_2019-2024.xlsx")

df = df[df["has_complete_trajectory"] == 1].copy()

# --- 1. Trajectory distribution ---
traj = df["trajectory_group"].value_counts(normalize=True)
print("\nTrajectory distribution:")
print(traj)

# --- 2. Trajectory × Sector ---
tab_sector = pd.crosstab(
    df["trajectory_group"],
    df["sector_en"],
    normalize="index"
)
print("\nTrajectory × Sector:")
print(tab_sector.round(3))

# --- 3. Trajectory × Ownership ---
ownership_col = "Foreign" if "Foreign" in df.columns else "owner_num"

tab_owner = pd.crosstab(
    df["trajectory_group"],
    df[ownership_col],
    normalize="index"
)
print("\nTrajectory × Ownership:")
print(tab_owner.round(3))

# --- 4. Growth by group ---
growth = df.groupby("trajectory_group")[
    ["growth_real_P1", "growth_real_P2", "growth_real_P3"]
].mean()
print("\nMean growth by group:")
print(growth.round(3))

median_table = df.groupby("trajectory_group")[
    ["growth_real_P1","growth_real_P2","growth_real_P3"]
].median()

print("\nMedian growth by group:")
print(median_table.round(3))
```
