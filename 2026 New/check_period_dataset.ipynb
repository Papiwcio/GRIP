```python
# -------------------------------------------------------
# DESCRIPTIVE DIAGNOSTICS
# -------------------------------------------------------
```

```python
growth_cols_real = ["growth_real_P1","growth_real_P2","growth_real_P3"]
growth_cols_log = ["growth_log_ann_P1","growth_log_ann_P2","growth_log_ann_P3"]

print("\nReal growth:")
display(df[growth_cols_real].describe(percentiles=[0.01,0.05,0.95,0.99]))

print("\nAnnualised log growth:")
display(df[growth_cols_log].describe(percentiles=[0.01,0.05,0.95,0.99]))
```

```python
print("\nData availability:")
print(df[["has_P1_data","has_P2_data","has_P3_data"]].sum())
```

```python
print("\nExtreme values (1% / 99%):")
display(df[growth_cols_real].quantile([0.01,0.99]))
```

```python
print("\nGrowth by manufacturing:")
display(df.groupby("manufacturing")[growth_cols_real].mean())

print("\nGrowth by ownership:")
display(df.groupby("owner")[growth_cols_real].mean())
```
