# DATA PREPARATION

# ANALYSIS  SET-UP

```python
import pandas as pd
import numpy as np
import re

INPUT_FILE = "Data_panel_2019-2024.parquet"
OUTPUT_XLSX = "Data_stat_2019-2024.xlsx"
OUTPUT_PARQUET = "Data_stat_2019-2024.parquet"


df = pd.read_parquet(INPUT_FILE)
```

# CATEGORIES

```python
# Avoid division by zero by converting 0 denominators to NaN
def safe_div(numerator, denominator):
    return numerator / denominator.replace({0: np.nan})

# Polish / Foreign Companies
df['owner'] = np.where((df['owner_type'].astype(str).str[:1] == '5'), 'Foreign', 'Domestic')
df['owner'] = df['owner'].astype('category')

# Pivot the DataFrame to make year-based comparison easier for both Sales and Export
#(use only with 2023 set as a filter)
pivot_sales = df.pivot(index='nip', columns='year', values='sales')
pivot_export_sales = df.pivot(index='nip', columns='year', values='exports')

# High Internationalisation Export Sales > 25% of Sales
pivot_sales['Hint25'] = (pivot_export_sales[2019] / pivot_sales[2019]) > 0.25
pivot_sales['Hint25'] = pivot_sales['Hint25'].apply(lambda x: 'Exp>25' if x else 'Exp<25')

# High Internationalisation Export Sales > 50% of Sales
pivot_sales['Hint50'] = (pivot_export_sales[2019] / pivot_sales[2019]) > 0.50
pivot_sales['Hint50'] = pivot_sales['Hint50'].apply(lambda x: 'Exp>50' if x else 'Exp<50')


# HGX Export Driven
pivot_sales['HGX'] = (pivot_export_sales[2024] / pivot_sales[2024]) > (pivot_export_sales[2019] / pivot_sales[2019])
pivot_sales['HGX'] = pivot_sales['HGX'].apply(lambda x: 'Increased Export Intensity' if x else 'Decreased Export Intensity')

pivot_sales['HGX_num'] = pivot_sales['HGX'].map({
    'Increased Export Intensity': 1,
    'Decreased Export Intensity': 0
})

# HGX Export Driven
pivot_sales['HGX%'] = (
    pivot_export_sales[2024] / pivot_sales[2024] - 
    pivot_export_sales[2019] / pivot_sales[2019]
).fillna(0)


# Merge the calculated columns back to the original df_filter DataFrame
df = df.merge(pivot_sales[['Hint25','Hint50','HGX','HGX_num','HGX%']], on='nip')
```

# RATIOS

```python
# -------------------------------------------------------
# FINANCIAL RATIOS + DUPONT DECOMPOSITION
# -------------------------------------------------------

# Avoid division by zero by converting 0 denominators to NaN
def safe_div(numerator, denominator):
    return numerator / denominator.replace({0: np.nan})

# ------------------------------
# BASIC FINANCIAL RATIOS
# ------------------------------

df["profit_margin"] = safe_div(df["net_profit"], df["sales"])

df["operating_margin"] = safe_div(df["operating_result"], df["sales"])

df["export_ratio"] = safe_div(df["exports"], df["sales"])

df["asset_turnover"] = safe_div(df["sales"], df["total_assets"])

df["capital_ratio"] = safe_div(df["equity"], df["total_assets"])

df["depreciation_ratio"] = safe_div(df["depreciation"], df["total_assets"])

df["wage_intensity"] = safe_div(df["wages_total"], df["sales"])


# ------------------------------
# DUPONT RATIOS
# ------------------------------

# Return on Assets
df["roa"] = safe_div(df["net_profit"], df["total_assets"])

# Return on Equity
df["roe"] = safe_div(df["net_profit"], df["equity"])

# Equity multiplier (financial leverage)
df["equity_multiplier"] = safe_div(df["total_assets"], df["equity"])


# ------------------------------
# EXTENDED DUPONT COMPONENTS
# ------------------------------

# Tax burden
df["tax_burden"] = safe_div(df["net_profit"], df["profit_before_tax"])

# Interest burden
df["interest_burden"] = safe_div(df["profit_before_tax"], df["operating_result"])


# ------------------------------
# PRODUCTIVITY VARIABLES
# ------------------------------

df["sales_per_employee"] = df["sales"] / df["employment"]
df["profit_per_employee"] = df["net_profit"] / df["employment"]
df["assets_per_employee"] = df["total_assets"] / df["employment"]
```

```python
# ------------------------------
# STARTING RATIOS
# ------------------------------

BASE_YEAR = 2019
PREFIX = "starting_"

# 1) Choose which columns should become "starting_" variables
#    (keep only columns that actually exist in df)
candidate_cols = [
    # levels
    "sales", "total_assets", "employment", "net_profit",
    
    # ratios (create ratios first, then include them here)
    "profit_margin", "operating_margin", "export_ratio",
    "asset_turnover", "capital_ratio", "equity_multiplier",
    "roa", "roe"
]

start_cols = [c for c in candidate_cols if c in df.columns]

# 2) Extract baseline year rows (one row per firm ideally)
start = (
    df.loc[df["year"] == BASE_YEAR, ["nip"] + start_cols]
      .sort_values("nip")
      .drop_duplicates(subset=["nip"], keep="first")
      .rename(columns={c: f"{PREFIX}{c}" for c in start_cols})
)

# 3) Merge starting variables back to the full panel
df = df.merge(start, on="nip", how="left")

# 4) Optional: add log starting size variables (common in regressions)
if "starting_total_assets" in df.columns:
    df["starting_ln_assets"] = np.log(df["starting_total_assets"].replace({0: np.nan}))
if "starting_sales" in df.columns:
    df["starting_ln_sales"] = np.log(df["starting_sales"].replace({0: np.nan}))
if "starting_employment" in df.columns:
    df["starting_ln_employment"] = np.log(df["starting_employment"].replace({0: np.nan}))

print("Created starting variables for base year:", BASE_YEAR)
print("Number of starting columns added:", len(start_cols))
```

# GROWTH

```python
# -------------------------------------------------------
# YEAR-ON-YEAR (YoY) GROWTH VARIABLES
# -------------------------------------------------------

# Always sort first for correct lagging
df = df.sort_values(["nip", "year"]).copy()

def yoy_growth(df, col, group="nip"):
    """
    YoY growth: (x_t / x_{t-1}) - 1
    Returns NaN if lag is missing or zero.
    """
    lag = df.groupby(group)[col].shift(1)
    return (df[col] / lag.replace({0: np.nan})) - 1

# Core YoY growth variables (add/remove as needed)
df["sales_growth_yoy"] = yoy_growth(df, "sales")
df["assets_growth_yoy"] = yoy_growth(df, "total_assets")
df["employment_growth_yoy"] = yoy_growth(df, "employment")
df["exports_growth_yoy"] = yoy_growth(df, "exports")

# Change rather than growth for profit (more stable)
df["net_profit_change_yoy"] = df["net_profit"] - df.groupby("nip")["net_profit"].shift(1)

# -------------------------------------------------------
# PERIOD GROWTH VARIABLES
# -------------------------------------------------------

# Keep only relevant years and create firm-level wide table
sales_wide = (
    df.loc[df["year"].isin([2019, 2023, 2024]), ["nip", "year", "sales"]]
      .drop_duplicates(subset=["nip", "year"])
      .pivot(index="nip", columns="year", values="sales")
      .rename(columns={
          2019: "sales_2019",
          2023: "sales_2023",
          2024: "sales_2024"
      })
      .reset_index()
)

# Replace zero starting sales with NaN to avoid division by zero
sales_wide["sales_2019_nonzero"] = sales_wide["sales_2019"].replace({0: np.nan})

# Absolute growth over period
sales_wide["sales_growth_2019_2023"] = (
    sales_wide["sales_2023"] / sales_wide["sales_2019_nonzero"] - 1
)

sales_wide["sales_growth_2019_2024"] = (
    sales_wide["sales_2024"] / sales_wide["sales_2019_nonzero"] - 1
)

# CAGR
sales_wide["sales_cagr_2019_2023"] = (
    (sales_wide["sales_2023"] / sales_wide["sales_2019_nonzero"]) ** (1/4) - 1
)

sales_wide["sales_cagr_2019_2024"] = (
    (sales_wide["sales_2024"] / sales_wide["sales_2019_nonzero"]) ** (1/5) - 1
)

# -------------------------------------------------------
# SALES VOLATILITY
# -------------------------------------------------------


sales_volatility = (
    df.groupby("nip")["sales_growth_yoy"]
      .std()
      .reset_index()
      .rename(columns={"sales_growth_yoy": "sales_growth_volatility"})
    
)

# Merge volatility into firm-level sales table
sales_wide = sales_wide.merge(sales_volatility, on="nip", how="left")

# -------------------------------------------------------
# SALES GROWTH CATEGORIES BASED ON 2019-2024 CAGR
# -------------------------------------------------------

# Thresholds in decimal form
inflation = 0.075
growth_threshold = 0.2

# Create explicit annualised growth variable
sales_wide["SGrowth_NR"] = sales_wide["sales_cagr_2019_2024"]

# Growth categories
#R2  – High growth
#R1  – Moderate real growth
#N1  – Low growth (below inflation)
#N2  – Decline

def categorize_growth1(row):
    g = row["SGrowth_NR"]
    
    if pd.isna(g):
        return np.nan
    elif g > growth_threshold:
        return "R2"
    elif inflation <= g <= growth_threshold:
        return "R1"
    elif 0 < g < inflation:
        return "N1"
    else:
        return "N2"

sales_wide["SGrowth_NR"] = sales_wide.apply(categorize_growth1, axis=1)

# Integrate
growth_vars = sales_wide[[
    "nip",
    "sales_growth_2019_2023",
    "sales_growth_2019_2024",
    "sales_cagr_2019_2023",
    "sales_cagr_2019_2024",
    "sales_growth_volatility",
    "SGrowth_NR"
]]

df = df.drop(columns=[col for col in growth_vars.columns if col != "nip" and col in df.columns], errors="ignore")
df = df.merge(growth_vars, on="nip", how="left")
```

```python
df.to_excel(OUTPUT_XLSX, index=False)
df.to_parquet(OUTPUT_PARQUET, index=False)
print("Saved files")
```
