```python
import pandas as pd

df = pd.read_excel("Data_period_2019-2024.xlsx")

df = df[
    (df["has_complete_trajectory"] == 1) &
    (df["manufacturing"] == 1) &
    (df["in_rank_2019"] == 1)
].copy()

# --- 1. Trajectory distribution ---
traj = df["trajectory_group"].value_counts(normalize=True)
print("\nTrajectory distribution:")
print(traj)

# --- 3. Trajectory × Ownership ---

ownership_col = "owner"  # "Foreign" / "Domestic"

# Counts
tab_owner_count = pd.crosstab(
    df["trajectory_group"],
    df[ownership_col]
)

# Add total row (counts)
tab_owner_count.loc["Total"] = tab_owner_count.sum()

# Row shares (%)
tab_owner_pct = pd.crosstab(
    df["trajectory_group"],
    df[ownership_col],
    normalize="index"
) * 100

# Add total row (% based on full sample)
total_pct = (
    tab_owner_count.loc["Total"] /
    tab_owner_count.loc["Total"].sum()
) * 100

tab_owner_pct.loc["Total"] = total_pct

print("\nTrajectory × Ownership - counts:")
print(tab_owner_count)

print("\nTrajectory × Ownership - row shares (%):")
print(tab_owner_pct.round(1))

# --- 4. Growth by group ---
growth = df.groupby("trajectory_group")[
    ["growth_real_P1", "growth_real_P2", "growth_real_P3"]
].mean()
print("\nMean growth by group:")
print(growth.round(3))

median_table = df.groupby("trajectory_group")[
    ["growth_real_P1","growth_real_P2","growth_real_P3"]
].median()

print("\nMedian growth by group:")
print(median_table.round(3))

# --- 4. Growth by N-R ---
growth = df.groupby("SGrowth_NR")[
    ["growth_real_P1", "growth_real_P2", "growth_real_P3"]
].mean()
print("\nMean growth by group:")
print(growth.round(3))

median_table = df.groupby("SGrowth_NR")[
    ["growth_real_P1","growth_real_P2","growth_real_P3"]
].median()

print("\nMedian growth by group:")
print(median_table.round(3))
```
